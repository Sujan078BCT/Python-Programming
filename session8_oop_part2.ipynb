### Write OOP classes to handle the following scenarios:

- A user can create and view 2D coordinates
- A user can find out the distance between 2 coordinates
- A user can find find the distance of a coordinate from origin
- A user can check if a point lies on a given line
- A user can find the distance between a given 2D point and a given line



````python
class Point:

  def __init__(self,x,y):
    self.x_cod = x
    self.y_cod = y

  def __str__(self):
    return '<{},{}>'.format(self.x_cod,self.y_cod)

  def euclidean_distance(self,other):
    return ((self.x_cod - other.x_cod)**2 + (self.y_cod - other.y_cod)**2)**0.5

  def distance_from_origin(self):
    return (self.x_cod**2 + self.y_cod**2)**0.5
    # return self.euclidean_distance(Point(0,0))


class Line:

  def __init__(self,A,B,C):
    self.A = A
    self.B = B
    self.C = C

  def __str__(self):
    return '{}x + {}y + {} = 0'.format(self.A,self.B,self.C)

  def point_on_line(line,point):
    if line.A*point.x_cod + line.B*point.y_cod + line.C == 0:
      return "lies on the line"
    else:
      return "does not lie on the line"

  def shortest_distance(line,point):
    return abs(line.A*point.x_cod + line.B*point.y_cod + line.C)/(line.A**2 + line.B**2)**0.5
````

````python
l1 = Line(1,1,-2)
p1 = Point(1,10)
print(l1)
print(p1)

l1.shortest_distance(p1)
````

### Reference Variables

- Reference variables hold the objects
- We can create objects without reference variable as well
- An object can have multiple reference variables
- Assigning a new reference variable to an existing object does not create a new object

````python
# object without a reference
class Person:

  def __init__(self):
    self.name = 'nitish'
    self.gender = 'male'

# creating object without reference variable
Person() # this create a instance of Person i.e object in memory location

# creating object with reference variable
p = Person() # p is not a object but a reference variable which contains/store memory address/location of object i.e points to that object
q = p
# now p and q points to same memory address where instance of Person is stored.
````

````python
# Multiple ref
print(id(p))
print(id(q))
````

````python
# change attribute value with the help of 2nd object
````

### Static Variables(Vs Instance variables)

**Needs of Static Variable**
- To implement counter

````python
# Trying to create unique customer Id.
class Atm:
  # constructor(special function)
  def __init__(self):
    self.pin = ''
    self.__balance = 0

    # Trying to create unique customer Id
    self.cid = 0
    self.cid += 1
````

````python
obj1 = Atm()
obj1.cid
````

````python
obj2 = Atm()
obj2.cid
````

**Problem**
- using instance variable , we cannot implement counter because cid for each object is independent i.e different i.e c1->cid c2->cid are independent and modifying c1->cid doesnot affect c2->cid
```
What is needed ?
- We need to have common cid(only one cid) for all object instead of having independent(different) for each.
i.e cid->c1,c2 and others.
- Modifying cid when performing operation in c1 object, it should be modified for other objects also.
```

**Solution**
- we need static variable.

**Two types of variables in class**
```
Instance variables
 -> unique value for each object
 -> It is object variables i.e belongs to object
 -> can be accessed using object only.
 -> declared under constructor.
```
```
Static variables
-> same value (common) for all object
->It is  class variables i.e belongs to class.
-> can be accessed by both object and class.
-> declared inside class but outside all method.
```

#####**Points to remember about static**

- **Static attributes are created at class level. i.e declared inside class but outside all method**
- **Static attributes are accessed using ClassName(mostly) but can also accessed through object.**
- **Static attributes are object independent. We can access them without creating instance (object) of the class in which they are defined.**
- **The value stored in static attribute is shared between all instances(objects) of the class in which the static attribute is defined.**

**When to use Static and Instance variables using examples ?**
```
You are designing Banking System
- Instance variable - customer_name,customer_account_no,customer_balance
- static variable - bank_name(because it is same for all customer)
```
```
You are designing college website
- Instance variable - student_name,student_cgpa(because different for each student)
- static variable - college_name,principal_name,degree_name(eg:bachelor)(because it is same for all student)
```

**How to access static attributes**
- Syntax

          className.static_attributes

          objectName.static_attributes

````python
class Person:
  person_id = 0 # defining public static attributes

  def __init__(self):
      self.id = Person.person_id + 1 # accessing public static attributes
      Person.person_id += 1 # modifying static attributes
````

**Problem**

````python
p1 = Person()
Person.person_id
````

````python
p2 = Person()
````

**Solution**
- Make static attributes also private
- create getter and setter method to access outside class.

````python
class Atm:

#defining static attributes as private
  __counter = 1

  def __init__(self):
    # defining instance attributes
    self.pin = ''
    self.__balance = 0
    self.cid = Atm.__counter # accessing private static attributes using class name
    Atm.__counter = Atm.__counter + 1

  def access(self):
    return self.__counter # accessing private static attributes using object

  # utility functions
  @staticmethod                 # Adding decorator
  def get_counter():  # since self is not used inside get_counter method , we can remove it.
    return Atm.__counter
````

### Static methods
- can access static(class) attributes only but not instance attributes because don't have self as parameter.
- can be accessed using both object and classname.
- Marked using the @staticmethod decorator.
- belongs to class and can be called without creating object.
- Used to create utility/helper functions, because they dont depend on object or class state and can be called with just the classname.

### Instance (non-static) methods
- used to access both static and instance attributes.
- can be accessed using object only.
- belongs to object and can be called with object only.


````python
class Lion:
  __water_source="well in the circus"

  def __init__(self,name, gender):
      self.__name=name
      self.__gender=gender

  def drinks_water(self):
      print(self.__name,
      "drinks water from the",Lion.__water_source)

  @staticmethod
  def get_water_source():
      # print(self.__name)
      return Lion.__water_source

simba=Lion("Simba","Male")
simba.drinks_water()
print("Water source of lions:",Lion.get_water_source())
print( "Water source of lions:",simba.get_water_source())
````
