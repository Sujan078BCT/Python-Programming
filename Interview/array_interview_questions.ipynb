````python
# 2. Check if an array is sorted
L = [1,2,3,4,5]

flag = True

for i in range(0,len(L)-1):
  if L[i] > L[i+1]:
    flag = False

if flag:
  print('sorted')
else:
  print('not sorted')
````

````python
# 3. Find Min/Max in a given array

L = [21,1,34,23,54,11,10]

max_val = L[0]

for i in L:
  if i > max_val:
    max_val = i

print(max_val)
````

````python
# 4. Find the first element to occur k times in an array

L = [1,1,2,3,4,4,5,5]

k = 2

d = {}

for i in L:
  if i in d:
    d[i] = d[i] + 1
  else:
    d[i] = 1

for i in d:
  if d[i] == k:
    print(i)
    break
````

````python
# 5. Find duplicates in an array
L = [1,1,2,3,4,4,5,5]

d = {}

for i in L:
  if i in d:
    d[i] = d[i] + 1
  else:
    d[i] = 1

for i in d:
  if d[i] > 1:
    print(i)


````

````python
# 6. Rotate array to left d items

L = [1,2,3,4,5]
rotate = 2

for i in range(rotate):
  temp = L[0]
  for j in range(0,len(L)-1):
    L[j] = L[j+1]
  L[len(L)-1] = temp

print(L)
````

````python
# 8. Find continous subarray with a given sum(given non-negative numbers)
# return the starting and ending index of the subarray
# return 1st subarray in case of multiple

# Method 1-> Time Complexity - O(n3) because sum(List) also takes one for loop
L = [1,22,13,7,9,11,10]
S = 16

for i in range(0,len(L)):
  subarray = []
  for j in range(i,len(L)):
    subarray.append(L[j])
    if sum(subarray) == S:
      print(subarray)
````

````python
# 9. Find element with left side smaller/right side greater in an array

# Method 1 - Time Complexity : O(n2)
L = [3,1,2,5,8,7,9]

for i in range(1,len(L)-1):
  flag = True

  for j in range(0,i):
    if L[j] > L[i]:
      flag = False

  for k in range(i+1,len(L)):
    if L[k] < L[i]:
      flag = False

  if flag:
    print(L[i])

````

````python
# Method2 - TimeComplexity - O(n2)
# max(L) and min(L) function takes one entire loop to find maximum value
L = [3,1,2,5,8,7,9]

for i in range(1,len(L)-1):
  if max(L[:i]) < L[i] < min(L[i+1:]):
    print(L[i])
````

````python
# Method3 - Time Complexity - O(n) => optimized code
L = [3,1,2,5,8,7,9]

max_arr = []
min_arr = []

max_val = L[0]
min_val = L[-1]

# create max_list
for i in L:
  if i>max_val:
    max_val = i
  max_arr.append(max_val)
# create min_list
for i in range(len(L)-1,-1,-1):
  if L[i] < min_val:
    min_val = L[i]

  min_arr.insert(0,min_val)
# perform comparison
for i in range(1,len(L)-1):
  if max_arr[i-1] < L[i] < min_arr[i+1]:
    print(L[i])
````

````python
# 10. Maximum sum subarray - Method1 (Time complexity - O(n2) )
L = [-2,4,7,-1,6,-11,14,3,-1,-6]

d = {}

for i in range(0,len(L)):
  subarray = []
  for j in range(i,len(L)):
    subarray.append(L[j])
    d[sum(subarray)] = subarray[:]
max_val = max(d.keys())

for i in d:
  if i == max_val:
    print(d[i])
````

````python
# 10. Maximum sum subarray - Method2 (Time Complexity - O(n)) -> Optimized method
# Kadane's Algorithm - 4 parameter(best_sum,curr_sum,curr_subarray,best_subarray)

# import sys
# best_sum = 0 It doesn't work for list containing negative no. only
# For this, either best_sum = L[0] or -sys.maxsize-1

L = [-2,4,7,-1,6,-11,14,3,-1,-6]
# L = [-2,-4,-4,-1,-3,-5,-6,-11,-14]

# best_sum = -sys.maxsize-1
best_sum = L[0]
curr_sum = 0
curr_subarray = []
best_subarray = []
for i in L:
  if curr_sum + i >= i :
    curr_subarray.append(i)
    curr_sum += i
  else:
    curr_sum = i
    # curr_subarray = [i]
    curr_subarray.clear()
    curr_subarray.append(i)
  if curr_sum > best_sum :
    best_sum = curr_sum
    best_subarray = curr_subarray[:]
  # print("Current : {} ,{}".format(curr_subarray,curr_sum))
  # print("Best : {} {}".format(best_subarray,best_sum))
print(best_subarray)
````
