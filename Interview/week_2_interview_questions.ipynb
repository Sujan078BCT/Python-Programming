**1.What is aliasing?**
- When 2 or more variable or name points to same memory location, it is called aliasing.
- Benefits: - optimize memory
- when need to create multiple variable with same value,instead of assigning same value to multiple variable(a=500,b=500),we perform(a = 500,b=a)
```
  a = 4
  b = a

  Both name a and b are pointing to same memory location where 4 is stored.
  We can make use of id function to view memory location.
```


````python
# aliasing concepts :
a = 4
b = a
````

````python
print(id(a))
print(id(b))
````

````python
c = b
````

````python
print(id(c))
````

**Benefits of Aliasing**
- techniques used to optimize memory
- Instead of creating different memory location for same value, it allows to create single memory location for different variable.
- -5 to 256 automatically create single memory location even if we perform
```
a = 5
b = 5
c = 5
```
- But rather than this value
```
a = 258
b = 258
c = 258
print(id(a))
print(id(b))
print(id(c))
```
This create a different memory location for same value, Here we can optimize memory by creating aliasing.

````python
a = 257
b = 257
c = 257
print(id(a))
print(id(b))
print(id(c))
````

**2. What is garbage collection?**
- When value are present in memory location but it is not used by any of variables, it results into wastage of memory in RAM, Inorder to optimize memory,there is concept of garbage collection
- Garbage collection is an internal program that automatically periodically checks unused memory and removes such memory.
- Garbage collector is a program that executes garbage collection
- python doesnot gives power to programmer to delete at memory level unlike c,c++(i.e In C,C++, we can write program to delete at memory level)

````python
a = 'DSMP 2022-23'
b = a
c = b
````

````python
# To find how many variable are pointing to same memory location where value are stored
import sys
sys.getrefcount('DSMP 2022-23')
````

````python
print(id(a))
del a
del b
del c
# Garbage collector remove the unused value.
print(id('DSMP 2022-23')) # this again create a new memory location for value 'DSMP 2022-23'
````
