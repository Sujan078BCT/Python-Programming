**1.What is aliasing?**
- When 2 or more variable or name points to same memory location, it is called aliasing.
- useful when working with immutable datatypes.
- Benefits: - optimize memory
- when need to create multiple variable with same value,instead of assigning same value to multiple variable(a=500,b=500),we perform(a = 500,b=a)
```
  a = 4
  b = a

  Both name a and b are pointing to same memory location where 4 is stored.
  We can make use of id function to view memory location.
```


````python
# aliasing concepts :
a = 4
b = a
````

````python
print(id(a))
print(id(b))
````

````python
c = b
````

````python
print(id(c))
````

- Garbage collector is a program that executes garbage collection

**3 What is mutability and why is it dangerous in certain scenarios?**




**Mutability**
 - ability of datatypes to be change at its memory location.
 - change -> mutable datatype
 - unchange -> immutable datatype
 - depend upon data types.
 - e.g: list is mutable whereas tuple is immutable.

````python
# mutable types - List
L = [1,2,3]
print(id(L))

L.append(4)
print(L)
print(id(L))
````

format:  **[memory address | stored value]**
```
L = [1,2,3]
meaning that
L ----->[0x100 | [1,2,3] ]

L.append(4)
meaning that at same address, list is modified.
L ----->[0x100 | [1,2,3,4] ]
```


````python
T = (1,2,3)
print(id(T))

T = T + (4,)

print(T)
print(id(T))
````

```
T = (1,2,3)
meaning that
T ----->[0x200 | (1,2,3)]

```

**4. What is Cloning?**
- creating a copy.
- useful when working with mutable datatypes(list,set,dictionary)

````python
def func(data):
  data.append(4)

a = [1,2,3]
func(a[:]) # passing clone
print(a)
````

````python
a = [1,2,3]
# cloning
b = a[:]
````

````python
id(a)
````

````python
id(b)
````

````python
b.append(4)
b
````

````python
a
````

````python
# problem
a = {'name':'nitish','age':33}
# aliasing
b = a

b['gender'] = 'male'

print(b)
print(a)
````

````python
# solution
a = {'name':'nitish','age':33}
# cloning
b = a.copy()

b['gender'] = 'male'

print(b)
print(a)
````

**5. Differentiate between deep and shallow copies?**

**Shallow copies**
- obtain when copy method is used.
- create copy of outer only but not inside.
- work only when dataitems are immutable types.
- used if dataitems is 1D incase of list.
- possible only with mutable datatypes (list,set,dictionary)
- not possible with immutable datatypes(tuple,string,int,float,complex,bool)

**Problem**
- dataitems inside mutable datatypes(list,set,dictionary) are created in new copy as reference but not as copy
```
L = [1,2,[4,5]]
b = L.copy()
# modifing [4,5] in b will also modify in list L.
b[-1][0] = 400
print(L)

o/p : [1,2,[400,5]]
```


````python
# shallow copy works fine here
a = [1,2,3]
# shallow
b = a.copy()

b.append(4)

print(a)
print(b)
````

````python
# shallow copy - problem when using mutable dataitems
a = [1,2,3,[4,5]]
b = a.copy()
print("id of a->{} and b->{} are different because of copy".format(id(a),id(b)))
print("But id of [4,5] in a ->{} and [4,5] in b ->{} is still same.".format(id(a[-1]),id(b[-1])))
print("Modifying [4,5] in b also modify [4,5] in a")
b[-1][0] = 400
print(b)
print(a)
````

**Solution to shallow copy -> Deep copy**
- create copy of outer as well as all inner dataitems.
- useful when nested items.
- Deep Copy create a copy of all dataitems inside mutable types unlike shallow copy
- copy module is required
```
import copy
L = [1,2,3,[4,5,9]]
D = copy.deepcopy(L)

D has copy of mutable dataitems present inside it also unlike shallow copy. i.e [4,5,9]

D[-1][0] = 400
print(D) # o/p: [1,2,3,[400,5,9]]
print(L) # o/p: [1,2,3,[4,5,9]]
```


````python
# [:] -> shallow copy
````

````python
a = [1,2,3,[4,5]]
b = a[:] # shallow copy
b[-1][0] = 700
print(b)
print(a)
````

````python
# Deep copy
import copy
a = [1,2,3,[4,5]]
b = copy.deepcopy(a)
print(id(a),id(b))
print(id(a[-1]),id(b[-1])) # dataitems are also in separate location
b[-1][0] = 400
print(b)
print(a)
````
