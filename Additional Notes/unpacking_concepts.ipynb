# 1. Dictionary Unpacking

### Dictionary Unpacking in Python (`**`)

**Dictionary unpacking** lets you extract or merge keyâ€“value pairs from one or more dictionaries using the `**` operator.

---

## 1ï¸âƒ£ Unpacking into another dictionary (merge dictionaries)

```python
a = {"x": 1, "y": 2}
b = {"y": 20, "z": 3}

c = {**a, **b}
print(c)
```

**Output**

```python
{'x': 1, 'y': 20, 'z': 3}
```

ðŸ“Œ **Important rule**

* If keys collide, **later dictionaries override earlier ones**.

---

## 2ï¸âƒ£ Adding / overriding values while unpacking

```python
user = {"name": "Sujan", "role": "student"}

updated = {**user, "role": "developer"}
print(updated)
```

```python
{'name': 'Sujan', 'role': 'developer'}
```

---

## 3ï¸âƒ£ Using dictionary unpacking in function calls

You can pass dictionary values as **keyword arguments**.

```python
def create_user(name, age, country):
    print(name, age, country)

data = {"name": "Sujan", "age": 22, "country": "Nepal"}

create_user(**data)
```

Equivalent to:

```python
create_user(name="Sujan", age=22, country="Nepal")
```

---

## 4ï¸âƒ£ Mixing normal arguments and unpacking

```python
extra = {"age": 22}

def info(name, age):
    print(name, age)

info("Sujan", **extra)
```

---

## 5ï¸âƒ£ Unpacking inside dictionary literals (real-world pattern)

Very common in **FastAPI**, **Django**, and **config handling**:

```python
default_config = {"debug": False, "timeout": 30}
env_config = {"timeout": 60}

config = {**default_config, **env_config}
```

âœ”ï¸ Clean
âœ”ï¸ Readable
âœ”ï¸ Immutable-safe (original dicts unchanged)

---

## 6ï¸âƒ£ Difference vs `dict.update()`

```python
a = {"x": 1}
b = {"y": 2}

a.update(b)   # mutates a
```

```python
c = {**a, **b}  # creates new dict
```

ðŸ“Œ Prefer unpacking when you want **immutability**.

---

## 7ï¸âƒ£ Common mistakes âŒ

### âŒ Unpacking non-dict

```python
{**[("a", 1), ("b", 2)]}   # TypeError
```

âœ”ï¸ Correct:

```python
dict([("a", 1), ("b", 2)])
```

---

## 8ï¸âƒ£ Python version note

* Dictionary unpacking with `**` â†’ **Python 3.5+**
* Multiple unpacking â†’ **Python 3.6+**

---

### Mental Model ðŸ§ 

```python
{**d1, **d2}
```

means:

> â€œTake all keyâ€“value pairs from `d1`, then overwrite/add with pairs from `d2`â€



### Tuple Unpacking in Python

**Tuple unpacking** means assigning multiple values from a tuple (or any iterable) to multiple variables **in one step**.

---

## 1ï¸âƒ£ Basic tuple unpacking

```python
point = (10, 20)
x, y = point

print(x)  # 10
print(y)  # 20
```

ðŸ“Œ Number of variables **must match** number of values.

---

## 2ï¸âƒ£ Unpacking without parentheses (very common)

```python
x, y, z = 1, 2, 3
```

Python automatically treats the right side as a tuple.

---

## 3ï¸âƒ£ Swapping variables (classic Python feature)

```python
a = 5
b = 10

a, b = b, a
```

No temp variable needed.

---

## 4ï¸âƒ£ Unpacking from other iterables (not just tuples)

```python
x, y = [3, 4]        # list
a, b = "hi"          # string
k, v = {"a": 1}.items().__iter__().__next__()
```

Any **iterable** works.

---

## 5ï¸âƒ£ Using `*` (extended unpacking)

```python
numbers = (1, 2, 3, 4, 5)

a, *middle, b = numbers

print(a)       # 1
print(middle)  # [2, 3, 4]
print(b)       # 5
```

ðŸ“Œ `*` collects **remaining items into a list**.

---

## 6ï¸âƒ£ Ignoring values (Python convention)

```python
data = (1, 2, 3, 4)

x, _, _, y = data
```

`_` means â€œI donâ€™t care about this valueâ€.

---

## 7ï¸âƒ£ Nested tuple unpacking

```python
record = ("Sujan", (22, "Nepal"))

name, (age, country) = record
```

---

## 8ï¸âƒ£ Unpacking in loops (very common in real code)

```python
pairs = [(1, "a"), (2, "b"), (3, "c")]

for num, char in pairs:
    print(num, char)
```

---

## 9ï¸âƒ£ Function returns & unpacking

```python
def stats():
    return 10, 20, 30

a, b, c = stats()
```

Python returns a **tuple by default**.

---

## ðŸ”Ÿ Common errors âŒ

### âŒ Too many / too few values

```python
x, y = (1, 2, 3)
# ValueError: too many values to unpack
```

âœ”ï¸ Fix:

```python
x, *rest = (1, 2, 3)
```

---

## Mental Model ðŸ§ 

```python
a, b, c = iterable
```

means:

> â€œIterate through `iterable` and assign values **one by one** from left to rightâ€

---

| Feature  | Tuple Unpacking | Dictionary Unpacking |
| -------- | --------------- | -------------------- |
| Operator | `,` and `*`     | `**`                 |
| Works on | Any iterable    | dict only            |
| Output   | variables       | new dict / kwargs    |



# 2. List Unpacking

### List Unpacking in Python

**List unpacking** is the same mechanism as tuple unpacking â€” Python just **iterates over the list** and assigns values to variables.

---

## 1ï¸âƒ£ Basic list unpacking

```python
nums = [10, 20]
a, b = nums

print(a)  # 10
print(b)  # 20
```

ðŸ“Œ Number of variables must match list length (unless using `*`).

---

## 2ï¸âƒ£ Unpacking without explicitly using a list

```python
a, b, c = [1, 2, 3]
```

Works the same as tuple unpacking.

---

## 3ï¸âƒ£ Extended unpacking (`*` operator)

```python
values = [1, 2, 3, 4, 5]

first, *middle, last = values

print(first)   # 1
print(middle)  # [2, 3, 4]
print(last)    # 5
```

ðŸ“Œ `*middle` always becomes a **list**.

---

## 4ï¸âƒ£ Using `*` to ignore values

```python
data = [100, 200, 300, 400]

_, second, *_, last = data
print(second, last)
```

---

## 5ï¸âƒ£ Unpacking in loops (very common)

```python
pairs = [[1, "a"], [2, "b"], [3, "c"]]

for num, char in pairs:
    print(num, char)
```

Works for lists of lists too.

---

## 6ï¸âƒ£ Nested list unpacking

```python
matrix = [[1, 2], [3, 4]]

[a, b], [c, d] = matrix
```

---

## 7ï¸âƒ£ Function arguments + list unpacking

### Passing list elements as positional arguments

```python
nums = [3, 5]

def add(x, y):
    return x + y

print(add(*nums))
```

Equivalent to:

```python
add(3, 5)
```

---

## 8ï¸âƒ£ Mismatch errors âŒ

### âŒ Too many values

```python
a, b = [1, 2, 3]
```

### âœ”ï¸ Fix

```python
a, *rest = [1, 2, 3]
```

---

## 9ï¸âƒ£ List unpacking vs slicing

```python
nums = [1, 2, 3, 4]

a, *b = nums
```

vs

```python
a = nums[0]
b = nums[1:]
```

Unpacking is **cleaner and faster to read**.

---

## Mental Model ðŸ§ 

```python
x, y, *rest = my_list
```

means:

> â€œTake items from the list left â†’ right and assign them one by oneâ€

---

## Quick Comparison

| Feature      | List Unpacking | Tuple Unpacking |
| ------------ | -------------- | --------------- |
| Syntax       | Same           | Same            |
| Mutability   | List mutable   | Tuple immutable |
| `*rest` type | List           | List            |



### `*` used in functions in Python (Complete but Simple)

The `*` symbol in functions is about **packing and unpacking positional arguments**.

---

## 1ï¸âƒ£ `*args` â†’ packing arguments

### When you **donâ€™t know how many arguments** will be passed

```python
def add(*args):
    print(args)
    return sum(args)

add(1, 2, 3, 4)
```

**Output**

```python
(1, 2, 3, 4)
```

ðŸ“Œ `args` is a **tuple**
ðŸ“Œ All extra positional arguments are **packed** into it

---

## 2ï¸âƒ£ Without `*args` (why it exists)

```python
def add(a, b):
    return a + b

add(1, 2, 3)  # âŒ TypeError
```

But:

```python
def add(*args):
    return sum(args)  # âœ…
```

---

## 3ï¸âƒ£ `*` when calling a function â†’ unpacking

```python
nums = [2, 3]

def multiply(a, b):
    return a * b

print(multiply(*nums))
```

Equivalent to:

```python
multiply(2, 3)
```

ðŸ“Œ List / tuple values are **unpacked** into positional arguments.

---

## 4ï¸âƒ£ Mixing normal parameters and `*args`

```python
def log(level, *messages):
    print(level, messages)

log("INFO", "Server started", "User logged in")
```

```python
level = "INFO"
messages = ("Server started", "User logged in")
```

ðŸ“Œ Normal parameters come **before** `*args`.

---

## 5ï¸âƒ£ `*` as a separator (keyword-only arguments)

This is a **very important concept** ðŸ‘

```python
def create_user(name, *, age, country):
    print(name, age, country)

create_user("Sujan", age=22, country="Nepal")   # âœ…
create_user("Sujan", 22, "Nepal")               # âŒ
```

ðŸ“Œ Anything **after `*` must be passed as keyword arguments**
ðŸ“Œ No positional arguments allowed after `*`

---

## 6ï¸âƒ£ Combining `*args` and `**kwargs`

```python
def show(*args, **kwargs):
    print(args)
    print(kwargs)

show(1, 2, 3, name="Sujan", role="Developer")
```

```python
args   -> (1, 2, 3)
kwargs -> {'name': 'Sujan', 'role': 'Developer'}
```

---

## 7ï¸âƒ£ Real-world example (FastAPI / Python APIs)

```python
def query_db(query, *params):
    cursor.execute(query, params)
```

Call:

```python
query_db("SELECT * FROM users WHERE age > ?", 18)
```

ðŸ“Œ Flexible argument handling

---

## 8ï¸âƒ£ Common mistakes âŒ

### âŒ Using `*args` as a list

```python
def f(*args):
    args.append(1)  # âŒ tuple has no append
```

âœ”ï¸ Fix:

```python
args = list(args)
```

# `*` as a separator â†’ **Keyword-Only Arguments**

## What does `*` mean here?

When `*` appears **alone** in a function parameter list, it means:

> **â€œStop accepting positional arguments here. Everything after this must be passed by keyword.â€**

---

## Basic example (your code)

```python
def create_user(name, *, age, country):
    print(name, age, country)
```

### Valid call âœ…

```python
create_user("Sujan", age=22, country="Nepal")
```

### Invalid call âŒ

```python
create_user("Sujan", 22, "Nepal")
```

Why?
Because `age` and `country` are **keyword-only**.

---

## Why Python needs this feature

### Problem WITHOUT keyword-only args

```python
def create_user(name, age, country):
    ...
```

```python
create_user("Sujan", 22, "Nepal")
```

âŒ This is valid â€” but **unclear**

* Is `22` age or something else?
* Easy to mess up argument order

---

### Solution: keyword-only arguments

```python
def create_user(name, *, age, country):
    ...
```

Now:

* `name` â†’ positional or keyword
* `age`, `country` â†’ **keyword-only**

This **forces clarity**.

---

## How Python parses arguments (IMPORTANT)

Python processes arguments in this order:

1. Positional-only (`/`)
2. Positional-or-keyword
3. `*` separator
4. Keyword-only
5. `**kwargs`

Your function:

```python
def create_user(name, *, age, country):
```

means:

| Parameter | Allowed               |
| --------- | --------------------- |
| `name`    | positional OR keyword |
| `age`     | keyword-only          |
| `country` | keyword-only          |

---

## Proof with error message

```python
create_user("Sujan", 22, country="Nepal")
```

Error:

```text
TypeError: create_user() takes 1 positional argument but 2 were given
```

Python is telling you:

> You tried to pass a positional argument **after `*`**

---

## `*` vs `*args` (CRITICAL difference)

### âŒ This is NOT `*args`

```python
def f(*, x, y):
    print(x, y)
```

Here:

* `*` does **NOT collect arguments**
* It only acts as a **separator**

---

### `*args` collects arguments

```python
def f(*args):
    print(args)
```

Here:

* `args` is a tuple of positional arguments

---

## Combining `*args` and keyword-only arguments

```python
def log(level, *messages, timestamp):
    print(level, messages, timestamp)
```

### Valid call âœ…

```python
log("INFO", "Started", "Connected", timestamp="10:30")
```

### Invalid âŒ

```python
log("INFO", "Started", "Connected", "10:30")
```

ðŸ“Œ Everything **after `*messages` is keyword-only**

---

## Why this is heavily used in real code

### 1ï¸âƒ£ Prevent breaking changes in APIs

```python
def connect(host, *, timeout=30, retries=3):
    ...
```

Later you can safely add:

```python
def connect(host, *, timeout=30, retries=3, ssl=True):
```

No existing code breaks.

---

### 2ï¸âƒ£ FastAPI / Framework design

```python
def Depends(dependency, *, use_cache=True):
    ...
```

Forces readable usage:

```python
Depends(get_db, use_cache=False)
```

---

### 3ï¸âƒ£ Avoid bugs from wrong ordering

```python
send_email("hello", to="a@x.com", cc="b@x.com")
```

Much safer than:

```python
send_email("hello", "a@x.com", "b@x.com")
```
