### Truthy and Falsy Values in Python

In Python, every value has an inherent boolean value, meaning it can be evaluated as either `True` or `False` in a boolean context (like an `if` statement or when used with `bool()`).

**Falsy Values:** These are values that evaluate to `False`.
*   `False` (the boolean value itself)
*   `None` (represents the absence of a value)
*   Zero of any numeric type: `0`, `0.0`, `0j` (complex numbers)
*   Empty sequences: `''` (empty string), `[]` (empty list), `()` (empty tuple), `range(0)`
*   Empty mappings: `{}` (empty dictionary)
*   Empty sets: `set()`

**Truthy Values:** All other values are considered truthy; they evaluate to `True`.
*   `True` (the boolean value itself)
*   Any non-zero number (e.g., `1`, `-5`, `3.14`)
*   Any non-empty sequence (e.g., `'hello'`, `[1, 2]`, `(1,)`)
*   Any non-empty mapping (e.g., `{'a': 1}`)

Understanding truthy and falsy values is crucial for writing concise and Pythonic conditional statements.

```python
# Falsy values in Python
print(f"bool(False): {bool(False)}")
print(f"bool(0): {bool(0)}")
print(f"bool(0.0): {bool(0.0)}")
print(f"bool(None): {bool(None)}")
print(f"bool(''): {bool('')}")
print(f"bool([]): {bool([])}")
print(f"bool(()): {bool(())}")
print(f"bool({{}}): {bool({})}) # Empty dictionary")
print(f"bool(set()): {bool(set())}")

print('\n--- Truthy Values ---\n')

# Truthy values in Python
print(f"bool(True): {bool(True)}")
print(f"bool(1): {bool(1)}")
print(f"bool('hello'): {bool('hello')}")
print(f"bool([1, 2]): {bool([1, 2])}")
print(f"bool((1,)): {bool((1,))}") # Non-empty tuple
print(f"bool({{'key': 'value'}}): {bool({'key': 'value'})}) # Non-empty dictionary")
print(f"bool(-5): {bool(-5)}")
```

### Bitwise NOT Operator (~) in Python

The bitwise NOT operator (`~`) in Python inverts all the bits of its operand. This means every `0` becomes `1` and every `1` becomes `0`. However, Python uses two's complement representation for signed integers. Due to this, the operation `~x` is equivalent to `-(x + 1)`.

Let's break down an example:

If we have the number `5`:
*   In binary (assuming 8-bit for simplicity, though Python handles arbitrary precision): `00000101`

When we apply the `~` operator:
*   `~5` inverts the bits: `11111010`

To find the decimal value of `11111010` in two's complement:
1.  Since the most significant bit is `1`, it's a negative number.
2.  Invert all bits again: `00000101`
3.  Add `1`: `00000110`
4.  The decimal value of `00000110` is `6`.
5.  So, the result is `-6`.

This confirms the formula `-(x + 1)`, as `-(5 + 1) = -6`.

### Bitwise NOT Operator (~) with Negative Numbers (8-bit Two's Complement Example)

Let's understand how `~-10` results in `9` using the 8-bit two's complement representation.

**1. Represent -10 in 8-bit Two's Complement:**
   *   Start with the positive equivalent, `10`:
       Binary of `10`: `0000 1010`
   *   Perform One's Complement (invert all bits):
       `1111 0101`
   *   Add `1` to get Two's Complement:
       `1111 0101 + 1 = 1111 0110`
   So, `-10` in 8-bit two's complement is `1111 0110`.

**2. Apply the Bitwise NOT operator (`~`) to -10:**
   This means inverting every bit of the two's complement representation of `-10`:
   `~(1111 0110) = 0000 1001`

**3. Interpret the Resulting Binary Value:**
   The result `0000 1001` is a positive number (since the most significant bit is `0`).
   Converting `0000 1001` from binary to decimal:
   ` (0 * 2^7) + (0 * 2^6) + (0 * 2^5) + (0 * 2^4) + (1 * 2^3) + (0 * 2^2) + (0 * 2^1) + (1 * 2^0)`
   `= 0 + 0 + 0 + 0 + 8 + 0 + 0 + 1`
   `= 9`

This example visually confirms the algebraic rule: `~x = -(x + 1)`. For `x = -10`,
`~(-10) = -(-10 + 1) = -(-9) = 9`.

```python
# Demonstrate ~-10 in Python
num = -10
print(f"Original number: {num}")
print(f"Bitwise NOT (~{num}): {~num}")
print(f"Check with formula -(num+1): {-(num + 1)}")
```

### Understanding `while...else` Blocks in Python

In Python, a `while` loop can optionally have an `else` block. The key behavior of this `else` block is:

*   **Execution Condition:** The `else` block associated with a `while` loop executes **only if the loop terminates normally**.
*   **Normal Termination:** A `while` loop terminates normally when its condition (the expression after `while`) becomes `False`.
*   **Skipped Execution:** If the `while` loop is terminated prematurely by a `break` statement, the `else` block is **skipped** and does not execute.

This behavior is useful for code where you want to perform an action after a loop completes successfully, without being interrupted.

```python
count = sum = 0
while True:
  num = int(input("Enter a number : "))
  count += 1
  sum += num
  if num == 0:
    break
else:  # else block is not executed  because while loop is terminated prematurely by a break statement
  print("control reaches.")
avg = sum / count
print(avg)
```
