## **Static vs Dynamic Typing**


**a. Static Typing**<br />
int a = 5; // variable are assigned with type, at time of creation
<br />
**b. Dynamic Typing**<br />
name = "Hari" // variable are not assigned with type but automatically known by interpreter based on its store valued.

## **Static Vs Dynamic Binding**
- based on binding between variable and datatype
- static binding => variable binded with datatype
- dynamic binding => variable not binded with datatype

**a. Static Binding**<br/>
- means once variable got its datatype, its datatypes remain fixed.
###### Examples
- int a = 5;
- a = "Hello world"; // got error because trying to assign string type to integer variable.

**b. Dynamic Binding**<br />
- means variable don't have fixed datatype.i.e keeps goes on changing based on stored value. <br/>
###### Examples
- a = 5 <br/>
- print(a) # 5<br />
- a = "hari"<br/>
- print(a) # hari <br/>

#### Two types of Type conversion:
- Implicit => automatically done by interpreter/compiler
- Explicit => done by programmer

```python
# Implicit Type Conversion
print(5 + 5.6) # print(float(5) + 5.6) automatically done by interpreter
print(type(5),type(5.6))
```
