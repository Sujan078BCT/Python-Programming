### Program - The current population of a town is 10000. The population of the town is increasing at the rate of 10% per year. You have to write a program to find out the population at the end of each of the last 10 years.

```python
# Code here
curr_pop = 10000

for i in range(10,0,-1):
  print(i,curr_pop)
  curr_pop = curr_pop/1.1
```

### Pattern 1

User enter  number of rows e.g row = 5 <br>
\* <br>
** <br>
*** <br>
**** <br>
*****<br>


```python
# code here

rows = int(input('enter number of rows'))

for i in range(1,rows+1):
  for j in range(1,i+1):
    print('*',end='')
  print()
```

### Pattern 2

1<br>
121<br>
12321<br>
1234321<br>

```python
# Code here
rows = int(input('enter number of rows'))

for i in range(1,rows+1):
  for j in range(1,i+1):
    print(j,end='')
  for k in range(i-1,0,-1):
    print(k,end='')

  print()
```

# Loop Control Statement

- Break
- Continue
- Pass

```python
for i in range(1,10):
  if i == 5:
    break
  print(i)
```

```python
lower = int(input('enter lower range'))
upper = int(input('enter upper range'))

for i in range(lower,upper+1):
  for j in range(2,i):
    if i%j == 0:
      break # can be used in linear search.
  else:
    print(i)
```

## Slicing
- Used to extract multiple characters.
- Syntax

[ startIndex , endIndex , stepsize and ordering ]
<br>
endIndex -> This position is excluded.
<br>
ordering -> positive => extracting in same order<br>
          -> Negative => extracting in reverse order
<br>Examples:
<br> [:] represent extracting or copying whole string
 <br>     [::] same as [:]
 <br>     [::-1] represent extracting whole string in reverse order
 <br>  **When even taking negative stepsize or reversing string, startIndex must be greater than endIndex**

```python
# Whenever trying to reverse string , StartIndex must be greater than endIndex
# Examples
name = "Ram Prasad"
# When printing reverse string by making endIndex greater than startIndex, it doesn't show result
print(name[0:3:-1])
# when startIndex greater than endIndex
print(name[4:0:-1])
```

```python
# Extract username from a given email.
# Eg if the email is nitish24singh@gmail.com
# then the username should be nitish24singh

email = input("Enter Email : ")
username = ""
for char in email :
  if char.isalnum():
    username += char
  else:
    break
print("username is {}".format(username))
```

```python
# Count the frequency of a particular character in a provided string.
# Eg 'hello how are you' is the string, the frequency of h in this string is 2.

userInput = input("Enter any String : ")
char = input("Enter any character : ")
# count = userInput.count(char)
count = 0
for i in userInput :
  if char == i :
    count = count + 1
print("Occurence of  {} in {} is {}".format(char,userInput,count))
```

```python
# Write a program which can remove a particular character from a string.
userInput = input("Enter any string : ")
char = input("Enter a character to remove : ")
removedString = ""
for new_char in userInput :
  if char != new_char :
    removedString += new_char
print("removed string is {}".format(removedString))
```

```python
# Write a program that can check whether a given string is palindrome or not.
# abba
# malayalam
userInput = input("Enter any string : ")
reversedInput = userInput[::-1]
if userInput == reversedInput :
  print("{} is palindrome.".format(userInput))
else :
  print("{} is not palindrome.".format(userInput))
```

```python
# Write a program to count the number of words in a string without split()
userInput = input("Enter any string : ")
count = 0
for i in userInput :
  if i == ' ':
    count = count + 1
count = count + 1
print("No. of words in a string is {}".format(count))
```

```python
# Write a python program to convert a string to title case without using the title()
userInput = input("Enter any string : ")
titlecasedString = ""
prevChar = ""
for i in userInput :
  currentChar = i
  if prevChar == "" or prevChar == " " :
    currentChar = i.upper()
  titlecasedString += currentChar
  prevChar = currentChar
print(titlecasedString)

```

```python
# Write a program which can remove a particular character from a string.
s = input('enter the string')
term = input('what would like to remove')

result = ''

for i in s:
  if i != term:
    result = result + i

print(result)
```

```python
# Write a program that can check whether a given string is palindrome or not.
# abba
# malayalam

s = input('enter the string')
flag = True
for i in range(0,len(s)//2):
  if s[i] != s[len(s) - i -1]:
    flag = False
    print('Not a Palindrome')
    break

if flag:
  print('Palindrome')

```

```python
# Write a program to count the number of words in a string without split()

s = input('enter the string')
L = []
temp = ''
for i in s:

  if i != ' ':
    temp = temp + i
  else:
    L.append(temp)
    temp = ''

L.append(temp)
print(L)

```

```python
# Write a python program to convert a string to title case without using the title()
s = input('enter the string')

L = []
for i in s.split():
  L.append(i[0].upper() + i[1:].lower())

print(" ".join(L))
```

```python
# Write a program that can convert an integer to string.

number = int(input('enter the number'))

digits = '0123456789'
result = ''
while number != 0:
  result = digits[number % 10] + result
  number = number//10

print(result)
print(type(result))
```
