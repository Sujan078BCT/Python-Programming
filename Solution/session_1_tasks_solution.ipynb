### Q2:- Write a program that will convert celsius value to fahrenheit.

````python
# Write your code here
# user input temperature in celcius
temp = float(input("Enter temperature in celcius degree : "))
# calculation
fahrenheit = (9/5)*temp + 32
print("Equivalent Fahrenheit temperature is : ",fahrenheit,"degree fahrenheit")
````

### Q3:- Take 2 numbers as input from the user.Write a program to swap the numbers without using any special python syntax.

````python
# Write your code here
# user input 2 integer number
num_1 = int(input("Enter first number : "))
num_2 = int(input("Enter second number : "))
print("Before Swapping : num_1 :",num_1,"num_2 :",num_2)
temp = num_1
num_1 = num_2
num_2 = temp
print("After Swapping : num_1 :",num_1,"num_2 :",num_2)
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Write your code here
# user input 2 co-ordinate
print("First Co-ordinate : ")
x1_x = float(input("Enter x-coordinate : "))
x1_y = float(input("Enter y-coordinate : "))
print("Second Co-ordinate : ")
x2_x = float(input("Enter x-coordinate : "))
x2_y = float(input("Enter y-coordinate : "))

x_distance = (x2_x - x1_x)**2
y_distance = (x2_y - x1_y)**2
distance = (x_distance + y_distance)**0.5

print("Distance : ",distance)
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.


````python
# Write your code here
# user input principal, rate and time period
principal = float(input("Enter principal : "))
rate = float(input("Enter rate : "))
time = int(input("Enter time period : "))
simple_interest = (principal * rate * time )/ 100
print("Simple Interest : ",simple_interest)
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here
# user input total no of heads and legs
head = int(input("Enter total no. of heads : "))
leg = int(input("Enter total no. of legs : "))
# performing calculation
dog = (leg // 2) - head
chicken = head - dog
print("No. of chicken and dog are :",chicken,"and",dog,"respectively.")
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Write your code here
# user input
number = int(input("Enter number : "))
# using formula
sum_of_square = (number*(number+1)*(2*number+1))/6
print("Sum of square is : ",sum_of_square)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Write your code here
# user input first 2 terms
a = int(input("Enter first term : "))
b = int(input("Enter second term : "))
n = int(input("Value of n : "))
nth_term = a + (n - 1)*(b - a)
print(n,"th term is :",nth_term)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
# Write your code here
# user input first and second fraction
print("First Fraction : ")
numerator_1 = int(input("Enter numerator : "))
denominator_1 = int(input("Enter denominator : "))
print("Second Fraction : ")
numerator_2 = int(input("Enter numerator : "))
denominator_2 = int(input("Enter denominator : "))
# calculation
num_1 = numerator_1 / denominator_1
num_2 = numerator_2 / denominator_2
sum = num_1 + num_2
print("Sum :",sum)
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# Write your code here
# User input milk tank and glass dimension
print("Milk tank Dimension : ")
tank_height = int(input("Enter milk tank height : "))
tank_width = int(input("Enter width of milk tank : "))
tank_breadth = int(input("Enter breadth of milk tank : "))
print("Glass Dimension : ")
glass_height = int(input("Enter glass height : "))
glass_radius = int(input("Enter glass radius : "))
# calculation
tank_volume = tank_height * tank_width * tank_breadth
glass_volume = 3.141597 * (glass_radius ** 2 ) * glass_height
number_of_glass = tank_volume // glass_volume
print("No. of glass :",number_of_glass)
````
