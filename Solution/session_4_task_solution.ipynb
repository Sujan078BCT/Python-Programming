###`Problem 1:` Combine two lists index-wise(columns wise)

Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list, then the 1st index item, and so on till the last element. any leftover items will get added at the end of the new list.

`Given List:`
```
list1 = ["M", "na", "i", "Kh"]
list2 = ["y", "me", "s", "an"]
```

`Output:`
```
[['M','y'], ['na', me'], ['i', 's'], ['Kh', 'an']]
```

````python
list1 = ["M", "na", "i", "Kh"]
list2 = ["y", "me", "s", "an"]

list3 = [[i,j] for (i,j) in zip(list1,list2)]
print(list3)
````

````python
# Write code here
list1 = ["M", "na", "i", "Kh","nam"]
list2 = ["y", "me", "s", "an"]
L = []
length = min(len(list1),len(list2))
# both list have index up to this
for i in range(length):
  list3 = []
  list3.extend([list1[i],list2[i]])
  L.append(list3)
# if extra element in list1
if length < len(list1):
  for i in range(length,len(list1)):
    L.append(list1[i])
# if extra element in list2
if length < len(list2):
  for i in range(length,len(list2)):
    L.append(list2[i])
print(L)
````

###`Problem 5:` You are given a list of integers. You are asked to make a list by running through elements of the list by adding all elements greater and itself.

i.e. Say given list is `[2,4,6,10,1]`
resultant list will be `[22,20,10,23]`.

For 1st element `2` ->> these are greater `(4+6+10)` values and `2` itself so on adding becomes `22`.

For 2nd element `4` ->> greater elements are `(6, 10)` and `4` itself, so on adding `20`

like wise for all other elememts.

`[2,4,6,10,1]`-->`[22,20,16,10,23]`




````python
# Write code here
list1 = [2,4,6,10,1]
L = []
for i in list1:
  sum = 0
  for j in list1:
    if i <= j:
      sum += j
  L.append(sum)
print(L)

# Time Complexity - O(n**2)
# Space Complexity - O(n)
````

###`Problem 6:` Find list of common unique items from two list. and show in increasing order

`Input`

```
num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]
```

`Output:`
```
[34, 67, 89]
```


````python
# Write code here
num1 = [23,45,67,78,89,34,34]
num2 = [34,34,89,55,56,39,67]
common = []
for item in num1:
  if item in num2 and item not in common:
    common.append(item)
output = sorted(common)
print(output)
````

````python
# Write code here
num1 = [23,45,67,78,89,34,34]
num2 = [34,89,55,56,39,67,34]
# may contain duplicate in result if same duplicate present in both num1 and num2
result = sorted([item for item in num1 if item in num2])
print(result)
````


###`Problem 7:` Sort a list of alphanumeric strings based on product value of numeric character in it. If in any string there is no numeric character take it's product value as 1.

`Input:`
```
['1ac21', '23fg', '456', '098d','1','kls']
```

`Output:`
```
['456', '23fg', '1ac21', '1', 'kls', '098d']

```

````python
# Write code here - incomplete answer
input = ['1ac21', '23fg', '456', '098d','1','kls']
product_list = []
for item in input:
  product = 1
  for char in item:
    if char.isdigit():
      product*=int(char)
  product_list.append(product)

output_list = [j for (i,j) in sorted(list(zip(product_list,input)),reverse=True)]
print(output_list)

````

### `Problem 10:` Add Space between Potential Words.

**Example:**

Input:

```bash
['campusxIs', 'bestFor', 'dataScientist']
```

Output:
```bash
['campusx Is', 'best For', 'data Scientist']
```

````python
# Code here

input = ['campusxIs', 'bestFor', 'dataScientist']
L = []
for item in input:
  result = ""
  for i in item:
    if i>='A' and i <='Z':
      result += ' '
    result += i
  L.append(result)
print(L)

````

### `Problem 11:` Write a program that can perform union operation on 2 lists

**Example:**

Input:

```bash
[1,2,3,4,5,1]
[2,3,5,7,8]
```

Output:
```bash
[1,2,3,4,5,7,8]
```

````python
# Write code here # Better Way
list1 = [1,2,3,4,5,1]
list2 = [2,3,5,7,8]
unique_list = []
for item in list1 + list2 :
  if item not in unique_list:
    unique_list.append(item)
output = sorted(unique_list)
print(output)
````

````python
list1 = [1,2,3,4,5,1]
list2 = [2,3,5,7,8]
union = []

for i in list1:
  if i not in list2:
    union.append(i)
for j in list2:
  if j not in union:
    union.append(j)
print(union)
````

````python
# Write code here
list1 = [1,2,3,4,5,1]
list2 = [2,3,5,7,8]
merge_list = list1 + list2
union = sorted([merge_list[i] for i in range(len(merge_list)) if merge_list[i] not in merge_list[i+1:]])
print(union)
````
