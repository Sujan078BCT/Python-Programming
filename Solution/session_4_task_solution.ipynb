###`Problem 1:` Combine two lists index-wise(columns wise)

Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list, then the 1st index item, and so on till the last element. any leftover items will get added at the end of the new list.

`Given List:`
```
list1 = ["M", "na", "i", "Kh"]
list2 = ["y", "me", "s", "an"]
```

`Output:`
```
[['M','y'], ['na', me'], ['i', 's'], ['Kh', 'an']]
```

````python
# Write code here
list1 = ["M", "na", "i", "Kh","nam"]
list2 = ["y", "me", "s", "an"]
L = []
length = min(len(list1),len(list2))
# both list have index up to this
for i in range(length):
  list3 = []
  list3.extend([list1[i],list2[i]])
  L.append(list3)
# if extra element in list1
if length < len(list1):
  for i in range(length,len(list1)):
    L.append(list1[i])
# if extra element in list2
if length < len(list2):
  for i in range(length,len(list2)):
    L.append(list2[i])
print(L)
````

###`Problem 3:` Update no of items available

Suppose you are given a list of candy and another list of same size representing no of items of respective candy.

i.e -  
```
candy_list = ['Jelly Belly','Kit Kat','Double Bubble','Milky Way','Three Musketeers']
no_of_items = [10,20,34,74,32]
```

Write a program to show no. of items of each candy type.

`Output:`

```
Jelly Belly-10
Kit Kat-20
Double Bubble-34
Milky Way-74
Three Musketeers-32

```

````python
# Write code here
candy_list = ['Jelly Belly','Kit Kat','Double Bubble','Milky Way','Three Musketeers']
no_of_items = [10,20,34,74,32]
for i,j in zip(candy_list,no_of_items):
  print("{}-{}".format(i,j))
````

###`Problem 5:` You are given a list of integers. You are asked to make a list by running through elements of the list by adding all elements greater and itself.

i.e. Say given list is `[2,4,6,10,1]`
resultant list will be `[22,20,10,23]`.

For 1st element `2` ->> these are greater `(4+6+10)` values and `2` itself so on adding becomes `22`.

For 2nd element `4` ->> greater elements are `(6, 10)` and `4` itself, so on adding `20`

like wise for all other elememts.

`[2,4,6,10,1]`-->`[22,20,16,10,23]`




````python
# Write code here
list1 = [2,4,6,10,1]
L = []
for i in range(len(list1)):
  sum = 0
  for j in range(0,len(list1)):
    if list1[i] <= list1[j]:
      sum += list1[j]
  L.append(sum)
print(L)
````

###`Problem 6:` Find list of common unique items from two list. and show in increasing order

`Input`

```
num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]
```

`Output:`
```
[34, 67, 89]
```


````python
# Write code here
num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]
result = [item for item in num1 if item in num2]
print(result)
````

### `Problem 10:` Add Space between Potential Words.

**Example:**

Input:

```bash
['campusxIs', 'bestFor', 'dataScientist']
```

Output:
```bash
['campusx Is', 'best For', 'data Scientist']
```

````python
# Code here

input = ['campusxIs', 'bestFor', 'dataScientist']
L = []
for item in input:
  result = ""
  for i in item:
    if i>='A' and i <='Z':
      result += ' '
    result += i
  L.append(result)
print(L)

````

### `Problem 11:` Write a program that can perform union operation on 2 lists

**Example:**

Input:

```bash
[1,2,3,4,5,1]
[2,3,5,7,8]
```

Output:
```bash
[1,2,3,4,5,7,8]
```

````python
# Write code here
list1 = [1,2,3,4,5,1]
list2 = [2,3,5,7,8]
merge_list = list1 + list2
union = [merge_list[i] for i in range(len(merge_list)) if merge_list[i] not in merge_list[i+1:]]
union.sort()
print(union)
````

### `Problem 14:` Write a list comprehension that can transpose a given matrix

matrix = [<br>

          [4,5,6],<br>
          [7,8,9]<br>]<br>

[1, 4, 7]<br>
[2, 5, 8]<br>
[3, 6, 9]<br>

````python
# Write code here
matrix = [[1,2,3],[4,5,6],[7,8,9]]
transpose = [[ matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix))]
print(transpose)
````
