### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# Write code here
# Method 1
# user input salary
salary = float(input("Enter salary in lakhs : "))
if salary < 5 :
  salary = salary * 0.82
elif salary < 10 :
  salary = salary * 0.72
elif salary < 20 :
  salary = salary * 0.62
else :
  salary = salary * 0.52
monthly_salary = salary / 12
print("Monthly salary : ",round(monthly_salary,2),"lakhs")
````

### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output:

```bash
120
```

````python
# Write code here
# user input
num = int(input("Enter integer number : "))
fact = 1
# perform calculation
for i in range(1,num+1) :
  fact = fact * i
print("Factorial of",num,"is",fact)
````

```bash
76542
```

`Output:`

```bash
24567
```

````python
# Write code here
# user input number
num = int(input("Enter integer number : "))
temp = num
reverse = 0
# perform calculation
while temp != 0 :
  remainder = temp % 10
  temp = temp // 10
  reverse = reverse * 10 + remainder
print("Reverse of",num,"is",reverse)
````

`Input:`

```bash

```

```bash
276
```

````python
# user input
num = int(input("Enter an integer number : "))
sum = 0
i = 1
while i <= num :
  if i % 5 != 0 :
    if sum + i > 300:
      break
    else :
      sum += i
  i += 1
print(sum)
````

### `Problem 9`: Write a program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numbers.

###`Problem 10`: Write a program, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number. The numbers obtained should be printed in a space-separated sequence on a single line.



````python
# Write code here
L = []
for num in range(1000,3001):
  is_even = True
  temp = num
  while temp != 0 :
    remainder = temp % 10
    temp //=10
    if remainder % 2 != 0 :
      is_even = False
      break
  if is_even :
    L.append(str(num))
print(','.join(L))
````

###`Problem 11`: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
from pickle import NONE
# Write code here
init_x_pos = init_y_pos = 0
final_y_pos = final_x_pos = 0
# user input direction
while True :
  user_input = input("Enter direction and movement : ")
  if user_input == '!':
    break
  direction = user_input.split(' ')[0].upper()
  movement = float(user_input.split(' ')[1])
  print(direction,movement)
  if direction == 'UP' :
    final_y_pos += movement
  elif direction == 'DOWN' :
    final_y_pos -= movement
  elif direction == 'LEFT' :
    final_x_pos -= movement
  elif direction == 'RIGHT' :
    final_x_pos += movement
  else :
    pass
distance = float((final_y_pos ** 2 + final_x_pos ** 2)**(1/2))
print("Distance : {}".format(distance))
````

###`Problem 12`:Write a program to print whether a given number is a prime number or not

````python
# Write code here
# user input integer number
num = int(input("Enter integer number : "))
is_prime = True
if num == 1 :
  is_prime = False
else :
  for i in range (2,num) :
    if num % i == 0:
      is_prime = False
      break
if is_prime :
  print("{} is a prime number.".format(num))
else :
  print("{} is not a prime number.".format(num))
````

###`Problem 13`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
# Write code here
# user input interger number
number = int(input("Enter integer number : "))
temp = number
sum = 0
while temp != 0:
  remainder = temp % 10
  temp //= 10
  sum += remainder ** 3
if sum == number :
  print("{} is armstrong number".format(number))
else :
  print("{} is not an armstrong number".format(number))
````
