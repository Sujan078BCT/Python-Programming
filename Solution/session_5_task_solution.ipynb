###`Q1:` Join Tuples if similar initial element
While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element. This problem can have applications in data domains such as Data Science.

For eg.
```
Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
Output : [(5, 6, 7, 8), (6, 10), (7, 13)]
```

````python
# Write code here -> Better Solution
test_list = [(5,6),(6,8),(6,10),(6,7),(7,13),(5,7)]
# find unique first index element in a set
unique = []
for i in test_list:
  unique.append(i[0])
unique = set(unique)

# merge same first element tuples
result = []
for i in unique:
  result.append([i])
  for j in test_list:
    if j[0] == i:
      result[-1].append(j[1])

# convert list of list into list of tuples
result = list(map(tuple,result))
result
````

````python
# write your code here - not a better way
test_list = [(5, 6), (5, 7), (5, 8), (6, 10),(7,13)]
visited_list = []
output_list = []
for i in range(len(test_list)):
  if test_list[i] in visited_list:
    continue
  visited_list.append(test_list[i])
  new_set = set()
  new_set.update(test_list[i])
  for j in range(i+1,len(test_list)):
    if test_list[i][0] == test_list[j][0]:
      new_set.update(test_list[j])
      visited_list.append(test_list[j])
  output_list.append(tuple(sorted(new_set)))
print(output_list)

````

###`Q2:` Multiply Adjacent elements (both side) and take sum of right and left side multiplication result.


For eg.
```
The original tuple : (1, 5, 7, 8, 10)
Resultant tuple after multiplication :

(1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)

output-(5, 40, 91, 136, 80)
```

````python
# Write code here
t = (1,5,7,8,10)

L = []
# handling first item
L.append(t[0]*t[1])

# handling middle item
for i in range(1,len(t)-1):
  L.append(t[i]*t[i-1] + t[i]*t[i+1])

# handling last item
L.append(t[-1]*t[-2])

print(tuple(L))
````

````python
# write your code here
tuples = (1,5,7,8,10)
output = []
for index in range(len(tuples)):
  if index == 0 :
    output.append(tuples[index]*tuples[index+1])
  elif index == len(tuples)-1 :
    output.append(tuples[index]*tuples[index-1])
  else:
    output.append(tuples[index]*(tuples[index-1] + tuples[index + 1]))
print("{}".format(tuple(output)))
````

###`Q4`: Count no of tuples, list and set from a list
```
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

```
`Output:`

```
List-2
Set-2
Tuples-1
```

````python
# write your code here
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]
item_dict = {"List":0,"Set":0,"Tuples":0}
for item in list1 :
  if type(item) == set:
    item_dict["Set"] = item_dict["Set"] + 1
  elif type(item) == tuple :
    item_dict["Tuples"] = item_dict["Tuples"] + 1
  elif type(item) == list :
    item_dict["List"] = item_dict["List"] + 1

for key in item_dict:
  print("{}-{}".format(key,item_dict[key]))
````

```
        ar2 = [6, 7, 20, 80, 100]

```

````python
# write your code here
ar1 = [1, 5, 10, 20, 40, 80]
ar2 = [6, 7, 20, 80, 100]
ar3 = [3, 4, 15, 20, 30, 70, 80, 120]
common_elem = []

for item in ar1 :
  if item in ar2 and item in ar3:
    common_elem.append(item)

# common_elem = [item for item in ar1 if item in ar2 and item in ar3]
print(common_elem)
````

````python
ar1 = [1, 5, 10, 20, 40, 80]
ar2 = [6, 7, 20, 80, 100]
ar3 = [3, 4, 15, 20, 30, 70, 80, 120]
# perform set intersection operation by casting all of 3 list into set
# common_elem = set(ar1) & set(ar2) & set(ar3)
common_elem = set(ar1).intersection(set(ar2)).intersection(set(ar3))
print("{}".format(list(common_elem)))
````

### `Q3:` Write a program to Check if a given string is binary string of or not.

A string is said to be binary if it's consists of only two unique characters.

Take string input from user.

```
Input: str = "01010101010"
Output: Yes

Input: str = "1222211"
Output: Yes

Input: str = "Campusx"
Output: No
```

````python
# write your code here
# solve using set cannot contain duplicate

input1 = input("Enter string : ")

input_set = set(input1) # this remove all duplicate

# if length of set == 2 -> yes else no

output = "Yes" if len(input_set) == 2 else "No"

print("Output : {}".format(output))
````

````python
#Question : check if string is binary number or not
# write code
input1 = input("Enter string : ")
binary_set = {'0','1'}
is_binary = True
for item in input1:
  if item not in binary_set:
    is_binary = False
    break
output = "Yes" if is_binary else "No"
print("Output : {}".format(output))
````

### `Q3`: Convert List to List of dictionaries. Given list values and keys list, convert these values to key value pairs in form of list of dictionaries.

**Example 1:**

Input:
```bash
test_list = ["DataScience", 3, "is", 8]
key_list = ["name", "id"]
```

Output:

```bash
[{'name': 'DataScience', 'id': 3}, {'name': 'is', 'id': 8}]
```

**Example 2:**

Input:
```bash
test_list = ["CampusX", 10]
key_list = ["name", "id"]
```

Output:

```bash
[{'name': 'CampusX', 'id': 10}]
```

````python
# write your code here
test_list = ["DataScience", 3, "is", 8]
key_list = ["name", "id"]

n = len(test_list)
res = []

for i in range(0,n,2):
  res.append({key_list[0]:test_list[i],key_list[1]:test_list[i+1]})

print(res)
````

````python
# write your code here
test_list = ["DataScience", 3, "is", 8]
key_list = ["name", "id"]
output_list = []
for index in range(0,len(test_list),2):
  output_dict = {}
  key_index = index
  for key in key_list:
    output_dict[key] = test_list[key_index]
    key_index = key_index + 1
  output_list.append(output_dict)
print(output_list)
````
