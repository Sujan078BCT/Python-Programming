### **`Problem-1:`** Write a Python function that takes a list and returns a new list with unique elements of the first list.

**Exercise 1:**

Input:

```bash
[1,2,3,3,3,3,4,5]
```

Output:

```bash
[1, 2, 3, 4, 5]
```

````python
input_list = [1,2,3,3,3,4,5]
def unique_element(input_list):
  unique_list = []
  for item in input_list:
    if item not in unique_list:
      unique_list.append(item)
  return unique_list
print(unique_element(input_list))
````

### **`Problem-2:`** Write a Python function that accepts a hyphen-separated sequence of words as parameter and returns the words in a hyphen-separated sequence after sorting them alphabetically.

**Example 1:**

Input:
```bash
green-red-yellow-black-white
```

Output:
```bash
black-green-red-white-yellow
```

````python
# Write code here
def sort_string(string):
  return '-'.join(sorted(string.split('-')))

input_string = "green-red-yellow-black-white"
output = sort_string(input_string)
print(output)
````

````python
# Write code here - Solution

def sort_sequence(seq):
  temp = []

  for i in sorted(seq.split('-')):
    temp.append(i)

  return '-'.join(temp)

s = 'green-red-yellow-black-white'
sort_sequence(s)
````

### **`Problem 3:`** Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.

```
Sample String : 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'
Expected Output :
No. of Upper case characters :  9
No. of Lower case Characters :  47
```

````python
# Write code here
Sample_String = 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'
def lower_upper(string):
  '''
  This function takes string and return list containing lowercase and uppercase count
  input : string
  output : list
  '''
  lowercase = uppercase = 0
  for item in string:
    if item.isupper():
      uppercase = uppercase + 1
    elif item.islower():
      lowercase = lowercase + 1
    else:
      pass
  return lowercase , uppercase

lowercase,uppercase = lower_upper(Sample_String)
print("No. of Upper case characters : {}".format(uppercase))
print("No. of lower case characters : {}".format(lowercase))

````

### **`Problem 5:`** Write a Python function to check whether a number is perfect or not.

A Perfect number is a number that is half the sum of all of its positive divisors (including itself).

Example :

```
The first perfect number is 6, because 1, 2, and 3 are its proper positive divisors, and 1 + 2 + 3 = 6.
Equivalently, the number 6 is equal to half the sum of all its positive divisors: ( 1 + 2 + 3 + 6 ) / 2 = 6.

The next perfect number is 28 = 1 + 2 + 4 + 7 + 14. This is followed by the perfect numbers 496 and 8128.
```


````python
# Write code here
user_input = int(input("Enter a number : "))
def is_perfect_number(num):
  '''
  This function takes integer number as input and return whether number is perfect or not.
  input : int number
  output : bool
  '''
  sum = 0
  for i in range(1,num):
    if num % i == 0:
      sum += i
  return sum == num
print(is_perfect_number(user_input))
````

### **`Problem-6:`** Write a Python function to concatenate any no of dictionaries to create a new one.

```
Sample Dictionary :
dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}
Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}
```

````python
# Write code here
def concate_dict(*args):
  output = {}
  for dictionary in args:
    output.update(dictionary)
  return output

dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}
print(concate_dict(dic1,dic2,dic3))
````

````python
# Write code here
def concate_dict(*args):
  output = {}
  for dictionary in args:
    for key in dictionary:
      output[key] = dictionary[key]
  return output

dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}
print(concate_dict(dic1,dic2,dic3))
````

`Problem-7` Write a python function that accepts a string as input and returns the word with most occurence.

```
Input:
hello how are you i am fine thank you
```

```
Output
you -> 2
```

````python
def most_occurence(input_string):
  # count occurence of each word
  output = {}
  for item in input_string.split():
    if item not in output:
      output[item] = 1
    else:
      output[item] += 1
  # find max occurence value
  max_occur = max(output.values())
  # find corresponding max value key
  for key in output:
    if output[key] == max_occur:
      print("{} -> {}".format(key,max_occur))
      break

most_occurence("hello how are you i am fine thank you")
````

`Problem-8` Write a python function that receives a list of integers and prints out a histogram of bin size 10

```
Input:
[13,42,15,37,22,39,41,50]
```

```
Output:
{11-20:2,21-30:1,31-40:2,41-50:3}
```

````python
# Write code here
import math
input = [10,13,42,15,37,22,39,41,50]
def create_histogram(L):
  min_bin = math.floor(min(L)/10)*10
  max_bin = math.ceil((max(L)/10))*10
  output = {}
  for val in range(min_bin,max_bin,10):
    count = 0
    for j in L:
      if val+1<= j <=val+10 :
        count += 1
    output["{}-{}".format(val+1,val+10)] = count

  return output

create_histogram(input)


````

**enumerate function**
- The enumerate() function is a built-in Python function that lets you loop through a list (or any iterable) while automatically keeping track of the index.
- creates enumerate object.
- use when need both index and value (basically while iterating)
- when typecast to list, it creates list of tuple(pos or index,item) indicating item index and item
- bascially used with iterables.
```
syntax:
enumerate(iterable, start=0)
- iterable → list, tuple, string, etc.
- start → the starting index (default = 0)
```
```
L = [0.909,1.2,2,4,3.9]
Inorder to remember position of each element when sorted or arrange
we make use of enumerate function

list(enumerate(L))
o/p: [(0,0.909),(1,1.2),(2,2.4),(3,4),(4,3.9)]

- We can perform sorting acc. to asc or desc remembering original position of list item but using sorted function.
```

````python
fruits = ["apple", "banana", "cherry"]
for i in range(len(fruits)):
  print(i,fruits[i])
# Cleaner, safer, and more Pythonic.
for index, fruit in enumerate(fruits):
    print(index, fruit)
````

````python
# With a custom starting index
for index, fruit in enumerate(fruits, start=1):
    print(index, fruit)
````

````python
# Write code here # using dictionary
import math
def shortest_dist(points,query):
  d = {}
  for i in points:
    distance = ((i[0]-query[0])**2 + (i[1]-query[1])**2)**0.5
    d[i] = distance

  dist_max = min(d.values())
  for i in d:
    if d[i] == dist_max:
      return i

coordinates = [(1,1),(2,2),(3,3),(4,4)]
shortest_dist(coordinates,(0,0))

````

`Problem-10`:Write a python program that receives a list of strings and performs bag of word operation on those strings

https://en.wikipedia.org/wiki/Bag-of-words_model

**Bag-of-words**
- algorithm in machine learning NLP to convert text into number.
```
input : list of string i.e ["cat mat bat","cat cat cat","bat bat bat"]
output : list of list i.e [[1,1,1,0],[3,0,0,0],[0,0,2,1]]

step1 : Find vocabulary i.e find all the unique word from all given sentences.(4 unique words -> cat mat bat rat)
1. cat mat bat
2. cat cat cat
3. bat bat rat
step2 : Convert to number by counting occurence of unique words in each sentences

Bag of Word representation for each sentences.
    cat mat bat rat
1. [1   1   1    0]
2. [3   0   0    0]
3. [0   0   2    1]
```

````python
# Write code here

def bow(L):
  # step 1 : find unique words in all given sentences
  vocab = set()
  for i in L:
    vocab.update(i.split()) # set union operation i.e s1 = s1 union s2 where s2 is iterables(set,tuple,list)

  # perform step 2
  result = []
  for i in L:
    result.append([])
    for j in vocab:
      result[-1].append(i.count(j))
  print(vocab)
  return result
L = [
    "cat mat rat cat",
    "sat bat fat cat rat",
    "pat cat mat rat"
]
bow(L)
````
